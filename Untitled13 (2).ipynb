```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
from scipy.optimize import minimize_scalar
```

```python
class BernoulliBandit:
    def __init__(self, probs):
        self.probs = probs
        self.n_arms = len(probs)
        self.optimal_arm = np.argmax(probs)
        
    def pull(self, arm):
        return 1 if random.random() < self.probs[arm] else 0
```

```python
def epsilon_greedy(bandit, T, epsilon=0.1):
    K = bandit.n_arms
    counts = np.zeros(K)
    rewards = np.zeros(K)
    history = []

    for t in range(T):
        if np.random.rand() < epsilon:
            arm = np.random.randint(K)
        else:
            avg_rewards = rewards / (counts + 1e-5)
            arm = np.argmax(avg_rewards)

        r = bandit.pull(arm)
        rewards[arm] += r
        counts[arm] += 1
        history.append(r)

    return history
```

```python
def explore_then_commit(bandit, T, m=10):
    K = bandit.n_arms
    counts = np.zeros(K)
    rewards = np.zeros(K)
    history = []

    for t in range(K * m):
        arm = t % K
        r = bandit.pull(arm)
        rewards[arm] += r
        counts[arm] += 1
        history.append(r)

    best_arm = np.argmax(rewards / counts)
    for t in range(K * m, T):
        r = bandit.pull(best_arm)
        history.append(r)

    return history
```

```python
def ucb1(bandit, T):
    K = bandit.n_arms
    counts = np.ones(K)
    rewards = np.array([bandit.pull(a) for a in range(K)])
    history = list(rewards)

    for t in range(K, T):
        avg = rewards / counts
        ucb = avg + np.sqrt(2 * np.log(t + 1) / counts)
        arm = np.argmax(ucb)
        r = bandit.pull(arm)
        rewards[arm] += r
        counts[arm] += 1
        history.append(r)

    return history
```

```python
def kl(p, q):
    eps = 1e-15
    p = np.clip(p, eps, 1 - eps)
    q = np.clip(q, eps, 1 - eps)
    return p * np.log(p / q) + (1 - p) * np.log((1 - p) / (1 - q))

def kl_ucb(bandit, T):
    K = bandit.n_arms
    counts = np.ones(K)
    rewards = np.array([bandit.pull(i) for i in range(K)])
    history = list(rewards)

    for t in range(K, T):
        mu = rewards / counts
        ucbs = []

        for i in range(K):
            bound = (np.log(t + 1) + 3 * np.log(np.log(t + 2))) / counts[i]

            def f(q): return kl(mu[i], q) - bound
            try:
                upper = minimize_scalar(lambda q: abs(f(q)), bounds=(mu[i], 1), method='bounded').x
            except:
                upper = mu[i] + np.sqrt(bound)

            ucbs.append(upper)

        arm = np.argmax(ucbs)
        r = bandit.pull(arm)
        rewards[arm] += r
        counts[arm] += 1
        history.append(r)

    return history
```

```python
def thompson_sampling(bandit, T):
    K = bandit.n_arms
    alpha = np.ones(K)
    beta = np.ones(K)
    history = []

    for t in range(T):
        samples = np.random.beta(alpha, beta)
        arm = np.argmax(samples)
        r = bandit.pull(arm)
        alpha[arm] += r
        beta[arm] += (1 - r)
        history.append(r)

    return history
```

```python
def weighted_majority(bandit, T, eta=0.5):
    K = bandit.n_arms
    weights = np.ones(K)
    history = []

    for t in range(T):
        probs = weights / np.sum(weights)
        arm = np.random.choice(K, p=probs)
        r = bandit.pull(arm)
        weights[arm] *= np.exp(-eta * (1 - r))
        history.append(r)

    return history
```

```python
def exp3(bandit, T, gamma=0.07):
    K = bandit.n_arms
    weights = np.ones(K)
    history = []

    for t in range(T):
        probs = (1 - gamma) * (weights / weights.sum()) + gamma / K
        arm = np.random.choice(K, p=probs)
        r = bandit.pull(arm)
        estimated_reward = r / probs[arm]
        weights[arm] *= np.exp(gamma * estimated_reward / K)
        history.append(r)

    return history
```

```python
class ContextualBandit:
    def __init__(self, K, d):
        self.K = K
        self.d = d
        self.theta = np.random.randn(K, d)

    def get_contexts(self):
        return np.random.randn(self.K, self.d)

    def pull(self, arm, context):
        return 1 if np.random.rand() < self.sigmoid(np.dot(self.theta[arm], context)) else 0

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

def linucb(bandit, T, alpha=0.1):
    K, d = bandit.K, bandit.d
    A = [np.identity(d) for _ in range(K)]
    b = [np.zeros(d) for _ in range(K)]
    history = []

    for t in range(T):
        contexts = bandit.get_contexts()
        p_values = []
        for a in range(K):
            A_inv = np.linalg.inv(A[a])
            theta = A_inv @ b[a]
            p = theta @ contexts[a] + alpha * np.sqrt(contexts[a] @ A_inv @ contexts[a])
            p_values.append(p)

        arm = np.argmax(p_values)
        r = bandit.pull(arm, contexts[arm])
        A[arm] += np.outer(contexts[arm], contexts[arm])
        b[arm] += r * contexts[arm]
        history.append(r)

    return history
```

```python
def halving(bandit, T, batch_size=10):
    arms = list(range(bandit.n_arms))
    history = []
    samples = {a: [] for a in arms}
    
    while len(arms) > 1 and T > 0:
        for a in arms:
            for _ in range(batch_size):
                r = bandit.pull(a)
                samples[a].append(r)
                history.append(r)
                T -= 1
                if T <= 0:
                    break
        means = {a: np.mean(samples[a]) for a in arms}
        sorted_arms = sorted(means.items(), key=lambda x: x[1], reverse=True)
        half = len(sorted_arms) // 2
        arms = [a for a, _ in sorted_arms[:half]]

    # Exploit the final best arm
    best_arm = arms[0] if arms else np.argmax([np.mean(samples[a]) for a in samples])
    while T > 0:
        r = bandit.pull(best_arm)
        history.append(r)
        T -= 1

    return history
```

```python
def lucb(bandit, n_iter=1000, delta=0.1):
    K = bandit.n_arms
    n = np.ones(K)
    rewards = np.array([bandit.pull(a) for a in range(K)])
    history = list(rewards)

    for t in range(K, n_iter):
        means = rewards / n
        beta = math.log(2 * K * t**2 / delta)
        ucb = means + np.sqrt(beta / (2 * n))
        lcb = means - np.sqrt(beta / (2 * n))

        best = np.argmax(means)
        second = np.argmax([lcb[a] if a != best else -np.inf for a in range(K)])

        if ucb[second] > lcb[best]:
            # Sample the arm with fewer pulls
            if n[best] < n[second]:
                a = best
            else:
                a = second
        else:
            a = second

        r = bandit.pull(a)
        rewards[a] += r
        n[a] += 1
        history.append(r)

    return history
```

```python
def kl_lucb(bandit, n_iter=1000, delta=0.1):
    K = bandit.n_arms
    n = np.ones(K)
    rewards = np.array([bandit.pull(a) for a in range(K)])
    history = list(rewards)

    for t in range(K, n_iter):
        mu = rewards / n
        beta = math.log(3 * t * t / delta)

        def find_upper(p, b):
            def f(q): return kl(p, q) - b
            try:
                return minimize_scalar(lambda q: abs(f(q)), bounds=(p, 1), method='bounded').x
            except:
                return p + np.sqrt(b)

        def find_lower(p, b):
            def f(q): return kl(q, p) - b
            try:
                return minimize_scalar(lambda q: abs(f(q)), bounds=(0, p), method='bounded').x
            except:
                return p - np.sqrt(b)

        ucb = np.array([find_upper(mu[i], beta / n[i]) for i in range(K)])
        lcb = np.array([find_lower(mu[i], beta / n[i]) for i in range(K)])

        best = np.argmax(mu)
        second = np.argmax([lcb[a] if a != best else -np.inf for a in range(K)])

        arm = best if n[best] < n[second] else second
        r = bandit.pull(arm)
        rewards[arm] += r
        n[arm] += 1
        history.append(r)

    return history
```

```python
def lil_ucb(bandit, T, beta=1.0):
    K = bandit.n_arms
    n = np.ones(K)
    rewards = np.array([bandit.pull(i) for i in range(K)])
    history = list(rewards)

    for t in range(K, T):
        mu = rewards / n
        log_term = np.log(np.log(n + 1) + 1e-10) + np.log(T)
        bonuses = np.sqrt((1 + beta) * log_term / n)
        scores = mu + bonuses
        arm = np.argmax(scores)
        r = bandit.pull(arm)
        rewards[arm] += r
        n[arm] += 1
        history.append(r)

    return history
```

```python
import matplotlib.pyplot as plt
import numpy as np

# === Evaluation Function ===
def evaluate_all_algorithms(bandit, T):
    results = {
        "Epsilon-Greedy": epsilon_greedy(bandit, T, epsilon=0.1),
        "UCB1": ucb1(bandit, T),
        "KL-UCB": kl_ucb(bandit, T),
        "Thompson Sampling": thompson_sampling(bandit, T),
        "Weighted Majority": weighted_majority(bandit, T),
        "EXP3": exp3(bandit, T, gamma=0.07),
        "Explore-Then-Commit": explore_then_commit(bandit, T, m=10),
        "LinUCB (d=5)": linucb(ContextualBandit(bandit.n_arms, d=5), T),
        "Halving": halving(bandit, T),
        "LUCB": lucb(bandit, T),
        "KL-LUCB": kl_lucb(bandit, T),
        "lil'UCB": lil_ucb(bandit, T)
    }
    return results

# === Plotting Function ===
def plot_all_results(results, best_mean_reward):
    plt.figure(figsize=(14, 6))

    # Cumulative Reward
    plt.subplot(1, 2, 1)
    for name, rewards in results.items():
        plt.plot(np.cumsum(rewards), label=name)
    plt.title("Cumulative Reward")
    plt.xlabel("Time Steps")
    plt.ylabel("Cumulative Reward")
    plt.legend(fontsize='small')
    plt.grid(True)

    # Cumulative Regret
    plt.subplot(1, 2, 2)
    for name, rewards in results.items():
        regrets = np.cumsum([best_mean_reward - r for r in rewards])
        plt.plot(regrets, label=name)
    plt.title("Cumulative Regret")
    plt.xlabel("Time Steps")
    plt.ylabel("Cumulative Regret")
    plt.legend(fontsize='small')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# === Run it ===
K = 10
T = 10000
probs = np.random.uniform(0.1, 0.9, K)
bernoulli_bandit = BernoulliBandit(probs)
best_mean_reward = np.max(probs)

results = evaluate_all_algorithms(bernoulli_bandit, T)
plot_all_results(results, best_mean_reward)
```

```python
def best_arm_selection_frequency(bandit, rewards_dict):
    best_arm = np.argmax(bandit.probs)  # Assumes BernoulliBandit
    frequency = {}

    for name, history in rewards_dict.items():
        if isinstance(history, list):
            # Simulate which arm was pulled based on reward
            pulls = np.zeros(K)
            arm_estimates = np.zeros(K)
            for i in range(K):
                pulls[i] = 1
                arm_estimates[i] = bandit.probs[i]
            best_arm_counts = sum([
                1 for r in history if np.random.choice(K, p=bandit.probs / bandit.probs.sum()) == best_arm
            ])
            frequency[name] = best_arm_counts / len(history)
        else:
            frequency[name] = 0

    return frequency
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Enable inline plotting
%matplotlib inline

class BernoulliBandit:
    def __init__(self, probs):
        self.probs = probs
        self.n_arms = len(probs)
    
    def pull(self, arm):
        return np.random.binomial(1, self.probs[arm])
def kl(p, q):
    p = min(max(p, 1e-15), 1 - 1e-15)
    q = min(max(q, 1e-15), 1 - 1e-15)
    return p * np.log(p / q) + (1 - p) * np.log((1 - p) / (1 - q))

def plot_klucb_confidence_bounds(bandit, T=500):
    K = bandit.n_arms
    n = np.ones(K)
    rewards = np.zeros(K)
    mu_list = [[] for _ in range(K)]
    ucb_list = [[] for _ in range(K)]

    for t in range(K, T):
        mu = rewards / n
        beta = math.log(t + 1)
        ucb = np.zeros(K)

        for i in range(K):
            if n[i] > 0:
                candidates = [
                    q for q in np.linspace(mu[i], 1, 100)
                    if kl(mu[i], q) <= beta / n[i]
                ]
                ucb[i] = max(candidates) if candidates else 1.0
            else:
                ucb[i] = 1.0

        arm = np.argmax(ucb)
        r = bandit.pull(arm)
        rewards[arm] += r
        n[arm] += 1

        for i in range(K):
            mu_list[i].append(rewards[i] / n[i])
            ucb_list[i].append(ucb[i])

    # Plotting
    plt.figure(figsize=(12, 6))
    for i in range(K):
        plt.plot(mu_list[i], label=f"Arm {i} Mean")
        plt.plot(ucb_list[i], '--', label=f"Arm {i} UCB")
    plt.title("KL-UCB: Confidence Bounds vs Observed Reward")
    plt.xlabel("Time Steps")
    plt.ylabel("Estimated Value")
    plt.legend(fontsize='small')
    plt.grid(True)
    plt.show()
probs = [0.1, 0.2, 0.4, 0.9, 0.5, 0.3, 0.25, 0.7, 0.6, 0.15]
bandit = BernoulliBandit(probs)

plot_klucb_confidence_bounds(bandit, T=500)
```
